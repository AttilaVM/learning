```python
from IPython.display import Audio
import numpy as np
```

```python
NOTE_DURATION = 1.0
SAMPLE_RATE = 22050
```

# Tunning for pentatonic scale

```python
ratios = [1, 3/2, 3/4, 3/2, 3/4]

root = 220
frequencies = []
for r in ratios:
    root = root * r
    frequencies.append(root)

tunes = sorted(frequencies)
tunes
```

```python
t = np.linspace(
    start=0, 
    stop=NOTE_DURATION,
    num=int(SAMPLE_RATE * NOTE_DURATION),
    endpoint=False
)
```

```python
audio_list = []

for tune in tunes:
    note = np.sin(tune * 2 * np.pi * t)
    note = note * (2**15 - 1) / np.max(np.abs(note))
    note = note.astype(np.int16)
    audio_list.append(note)
    
```

```python
audio = np.concatenate(audio_list)
```

```python
Audio(audio, rate=SAMPLE_RATE)
```

# Generalize playing

```python
def freq_to_audio(frequencies):
    
    t = np.linspace(
    start=0, 
    stop=NOTE_DURATION,
    num=int(SAMPLE_RATE * NOTE_DURATION),
    endpoint=False
    )
    
    audio_list = []

    for f in frequencies:
        note = np.sin(f * 2 * np.pi * t)
        note = note * (2**15 - 1) / np.max(np.abs(note))
        note = note.astype(np.int16)
        audio_list.append(note)
        
    audio = np.concatenate(audio_list)
    
    Audio(audio, rate=SAMPLE_RATE)
    
    return(audio)
    
Audio(freq_to_audio(tunes), rate=SAMPLE_RATE)
```

# Pythagorean tunning

```python
from fractions import Fraction
```

```python
r = Fraction(1)
ratios_low = [r]
for i in range(5):
    r = r *  Fraction(2, 3) * Fraction(2, 1)
    if r > Fraction(1, 2):
        ratios_low.append(r)
    else:
        r = r * Fraction(1, 2)
        ratios_low.append(r)
    i = i + 1
    
ratios_low.reverse()
ratios_low
```

```python
r = Fraction(1)
ratios = [r]
for i in range(8):
    r = r *  Fraction(3, 2)
    if r < 2:
        ratios.append(r)
    else:
        r = r * Fraction(1, 2)
        ratios.append(r)
    i = i + 1
    
ratios
```

```python
sorted([float(r) for r in ratios])
```

```python
freqs = [220 * float(r) for r in ratios]
freqs = sorted(freqs)
freqs
```

```python
Audio(freq_to_audio(freqs), rate=SAMPLE_RATE)
```

# Equal temeprament tunning

The base fraction of 2 (octave) is $\sqrt[12](2)$

```python
b = pow(2, 1/12)
fractions = []
for i in range(0, 13):
    r = b ** i
    fractions.append(r)
fractions
```

```python
frequencies = [220 * frac for frac in fractions]
frequencies
```

```python
Audio(freq_to_audio(frequencies), rate=SAMPLE_RATE)
```
